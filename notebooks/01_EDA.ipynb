```python
df_train = pd.read_csv('../artifacts/data/train.csv')
```

```python
# plt.figure(figsize=(24, 10))
_ , axes = plt.subplots(2, 3, figsize=(20, 15))
sns.set(style="whitegrid")

sns.countplot(data=df_train, x='housing', order=df_train['housing'].value_counts().index, ax=axes[0,0])
plt.xticks(rotation=45)
sns.countplot(data=df_train, x='previous', order=df_train['previous'].value_counts().index, ax=axes[0,1])
plt.xticks(rotation=45)
sns.countplot(data=df_train, x='loan', order=df_train['loan'].value_counts().index, ax=axes[1,0])
plt.xticks(rotation=45)
sns.countplot(data=df_train, x='job', order=df_train['job'].value_counts().index, ax=axes[1,2])
sns.countplot(data=df_train, x='contact', order=df_train['contact'].value_counts().index, ax=axes[1,1])
sns.countplot(data=df_train, x='poutcome', order=df_train['poutcome'].value_counts().index, ax=axes[0,2])

plt.xticks(rotation=45)
plt.title('Distribution of Job Types')
plt.show()
```

```python
poutcome_rate = df_train.groupby('poutcome')['y'].mean()

poutcome_rate.plot(kind='barh', figsize=(10, 6))
plt.title('Subscription Rate by PREVIOUS Outcome')
plt.xlabel('Subscription Rate (%)')
plt.show()
```

```python
job_rate = df_train.groupby('job')['y'].mean().sort_values()

job_rate.plot(kind='barh', figsize=(10, 8))
plt.title('Subscription Rate by Job')
plt.xlabel('Subscription Rate (%)')
plt.show()
#   'blue-collar', 'technician', 'services', 
#             'management', 'retired', 'student', 'self-employed'
# blue-collar      7441
# technician       5400
# services         3196
# management       2345
# retired          1366
# entrepreneur     1160
# self-employed    1099
# housemaid         855
# unemployed        798
# student           711
# unknown           265
```
